```python
# Load libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
url = "https://raw.githubusercontent.com/Group4-s2025/DATA6550-Bias/refs/heads/main/data/heart_attack_prediction_dataset.csv"
df = pd.read_csv(url)
```

```python
df.info()
```

```python
df.head()
```

# Sampling Bias: Gender Distribution

```python
# Gender distribution
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Sex")
plt.title("Sampling Bias: Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

# Print percentage distribution
gender_counts = df["Sex"].value_counts(normalize=True) * 100
print("Gender Distribution:\n", gender_counts)
```

### Explanation
If we train a predictive model using this dataset, the model may learn to predict patterns that are more reflective of male patients, considering they make up 70% of the data.
This means that the model could be inaccurate when predicting heart attack risk for female patients since there are less examples for the model to learn from.

# Selection Bias: Income

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Income"], bins=10, kde=True, color="purple")
plt.title("Selection Bias: Income Distribution")
plt.xlabel("Income Level")
plt.ylabel("Frequency")
plt.show()
```

### Explanation



# Historical Bias: Heart Attack Risk by Gender

```python
# Group by gender and calculate heart attack risk proportions
gender_heart_risk = df.groupby("Sex")["Heart Attack Risk"].mean()

# Bar plot
plt.figure(figsize=(6,4))
sns.barplot(x=gender_heart_risk.index, y=gender_heart_risk.values)
plt.title("Historical Bias: Heart Attack Risk by Gender")
plt.ylabel("Proportion with Heart Attack Risk")
plt.show()
```

### Explanation

# Geographic Bias: Continent & Country Distribution

```python
# Continent Distribution
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Continent")
plt.title("Geographic Bias: Continent Distribution")
plt.xticks(rotation=45)
plt.ylabel("Count")
plt.show()
```

```python
# Top 10 countries
top_countries = df["Country"].value_counts().nlargest(10)

# Bar plot for top countries
plt.figure(figsize=(8,5))
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.title("Geographic Bias: Top 10 Countries in Dataset")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

## Measuring Bias: Age Distribution


```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True, bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

## BMI Bias

```python
plt.figure(figsize=(12, 8))

# Histogram overlay
sns.histplot(df['BMI'], kde=False, bins=30, alpha=0.6)

plt.title('BMI Distribution with Frequency')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.show()
```

## Physical Activity Bias

```python
plt.figure(figsize=(12, 8))

# Violin plot for physical activity days per week segmented by smoking levels
sns.violinplot(x='Smoking', y='Physical Activity Days Per Week', data=df, inner='quartile')

plt.title('Physical Activity Levels by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Physical Activity Days Per Week')
plt.show()
```

## Sleeping Hours Bias with Heart Rate

```python
plt.figure(figsize=(12, 8))

# Violin plot for sleeping hours per day segmented by heart rate
sns.barplot(x='Sleep Hours Per Day', y='Heart Rate', data=df)

plt.title('Heart Rate Distribution by Sleep Hours')
plt.xlabel('Sleep Hours Per Day')
plt.ylabel('Heart Rate')
plt.show()
```

## Diabetics Bias

```python
# Assuming you have a DataFrame 'df' with the relevant data
plt.figure(figsize=(10, 6))

# Countplot for diabetes status
sns.countplot(x='Diabetes', data=df)
plt.title('Count of Diabetes Status')
plt.xlabel('Diabetes Status')
plt.ylabel('Count')
plt.show()
```

```python
plt.subplot(2, 3, 1)
sns.histplot(x='Diabetes', y='Cholesterol', data=df)
plt.title('Cholesterol Levels by Diabetes Status')
plt.xlabel('Diabetes Status')
plt.ylabel('Cholesterol Levels')
plt.show()
```

```python
# Violin plot for Heart Rate segmented by Diabetes Status
plt.subplot(2, 3, 3)
sns.boxplot(x='Diabetes', y='Heart Rate', data=df)
plt.title('Heart Rate by Diabetes Status')
plt.xlabel('Diabetes Status')
plt.ylabel('Heart Rate')
plt.show()
```

```python
# Violin plot for Physical Activity Days Per Week segmented by Diabetes Status
plt.subplot(2, 3, 4)
sns.barplot(x='Diabetes', y='Physical Activity Days Per Week', data=df)
plt.title('Physical Activity Days by Diabetes Status')
plt.xlabel('Diabetes Status')
plt.ylabel('Physical Activity Days Per Week')
plt.show()
```

```python
# Count plot for Diabetes Status
plt.subplot(2, 3, 5)
sns.countplot(x='Diabetes', data=df)
plt.title('Count of Diabetes Status')
plt.xlabel('Diabetes Status')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```
