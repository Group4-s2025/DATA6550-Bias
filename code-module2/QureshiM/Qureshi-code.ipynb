```python
df['Exercise Hours Per Week'].hist()
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of weekly exercise hrs')
plt.show()
```

```python
def exercise_label(exer):
    if exer < 0.5:
        return 'none'
    elif 0.5 <= exer < 2.5:
        return "1-2 "
    elif 2.5 <= exer < 5.5:
        return "3-5"
    elif 5.5 <= exer < 9.5:
        return "6-9"
    elif 9.5 <= exer < 15.5:
        return "10-15"
    elif 15.5 <= exer < 20.5:
        return "15-20"
    else:
        return "over 20"

df['Exercise_Label'] = df['Exercise Hours Per Week'].apply(exercise_label)
print(df['Exercise_Label'].value_counts())

frequencies = df['Exercise_Label'].value_counts()



plt.bar(frequencies.index, frequencies.values)


plt.xlabel('Exercise weekly hours')
plt.ylabel('Frequency')
plt.title('')


plt.show()
```

```python
df['BMI'].hist()
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of BMI')
plt.show()
```

```python
def bmi_label(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return "norm"
    elif 25 <= bmi < 30:
        return "Overweight"
    else:
        return "Obese"

df['BMI_Label'] = df['BMI'].apply(bmi_label)
print(df['BMI_Label'].value_counts())

frequencies = df["BMI_Label"].value_counts()



plt.bar(frequencies.index, frequencies.values)


plt.xlabel('BMI classifiers')
plt.ylabel('Frequency')
plt.title('Imbalanced BMI')


plt.show()
```

```python
def wealth(inc):
    if inc < 50000:
        return '<50k'
    elif 50000 <= inc < 100000:
        return "50k-100k"
    elif 50000 <= inc < 100000:
        return "50k-100k"
    elif 100000 <= inc < 150000:
        return "100k-150k"
    elif 150000 <= inc < 200000:
        return "150k-200k"
    elif 200000 <= inc < 250000:
        return "200k-250k"
    elif 250000 <= inc < 300000:
        return "250k-300k"
    else:
        return ">300k"

    
df['Income_Label'] = df['Income'].apply(wealth)
print(df['Income_Label'].value_counts())

frequencies = df["Income_Label"].value_counts()



plt.bar(frequencies.index, frequencies.values)


plt.xlabel('income')
plt.ylabel('Frequency')
plt.title('')


plt.show()
print(frequencies)
```
